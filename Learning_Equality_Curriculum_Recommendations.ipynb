```python
import zipfile
import pandas as pd
```

```python
local_zip = ['learning-equality-curriculum-recommendations.zip']
for i in range(len(local_zip)):
    zip_ref = zipfile.ZipFile(local_zip[i], 'r')
    zip_ref.extractall('/tmp')
    zip_ref.close()
```
